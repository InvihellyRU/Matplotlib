```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import matplotlib.patheffects as PathEffects
from matplotlib.ticker import AutoMinorLocator
from matplotlib.gridspec import GridSpec
from matplotlib import colormaps

from sklearn.datasets import load_iris
```

Диаграмма, изображающая значения двух переменных в виде точек на плоскости.

plt.scatter(x, y, data = '...', label=name, edgecolor='color', linewidth=int(), alpha = int())

plt.xlabel('...')
plt.ylabel('...')

Функция plt.scatter() принимает только два аргумента за раз - координаты x и y для одной группы точек. Чтобы задать разные координаты для каждой группы, нужно вызвать plt.scatter для каждой группы отдельно.

Когда вы вызываете plt.scatter() или любую другую функцию для построения графика без указания цвета, Matplotlib автоматически выбирает цвет для каждого нового графика из стандартной палитры цветов. Эта палитра определяет последовательность цветов, которые будут использоваться для различения разных наборов данных на графике.
Таким образом, каждый вызов plt.scatter() в вашем коде создает новый набор точек, и Matplotlib автоматически назначает каждому из них следующий цвет из палитры, чтобы графики были легко различимы. Вы также можете явно задать цвет, используя параметр color или c внутри функции scatter()

```python
iris_dataset = load_iris()
df = pd.DataFrame(iris_dataset['data'], columns=iris_dataset['feature_names'])
df['target'] = iris_dataset['target']
df['target'] = df['target'].map({0: 'setosa', 1: 'versicolor', 2: 'virginica'})

display(df)
```

```python
# 1 способ
x = df['petal length (cm)']
y = df['petal width (cm)']

plt.scatter(x, y)

plt.xlabel('Длина лепестка')
plt.ylabel('Ширина лепестка')

plt.show()

#---------------------------------------------

# 2 способ

plt.scatter(x='petal length (cm)', y='petal width (cm)', data=df)

plt.xlabel('Длина лепестка')
plt.ylabel('Ширина лепестка')

plt.show()
```

```python
x = 'petal length (cm)'
y = 'petal width (cm)'

groups = df.groupby('target')

for name, group in groups:
    plt.scatter(group[x], group[y], 
                label=name, 
                edgecolor='black', 
                linewidth=0.6, 
                alpha = 0.7)

plt.legend()
plt.xlabel('Длина лепестка')
plt.ylabel('Ширина лепестка')

plt.show()
```

```python
x = df['petal length (cm)']
y = df['petal width (cm)']
z = df['petal width (cm)'] / df['petal length (cm)']

plt.scatter(x, y, c=z, 
            cmap='viridis')

plt.xlabel('Длина лепестка')
plt.ylabel('Ширина лепестка')

cbar = plt.colorbar()
cbar.set_label('Отношение ширины лепестка к его длине')

plt.show()
```
