Столбчатая диаграмма (bar plot)

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import matplotlib.patheffects as PathEffects
from matplotlib.ticker import AutoMinorLocator
from matplotlib.gridspec import GridSpec
from matplotlib import colormaps
```

```python
df_values = [['2019', 7238, 7239],
             ['2020', 7497, 7840],
             ['2021', 8997, 8410],
             ['2022', 9472, 9107]]
df_columns = ['Год', 'Торговые площади Магнит, тыс. кв.м.',
              'Торговые площади X5Group, тыс. кв.м.']
df = pd.DataFrame(df_values, columns=df_columns)
```

```python
plt.figure(figsize=(8, 6))

ax = plt.bar(df['Год'], df['Торговые площади X5Group, тыс. кв.м.'], 
             width=0.4, 
             color='green', 
             alpha=0.6)

plt.title('Торговые площади X5Group')
plt.xlabel('Год')
plt.ylabel('Торговые площади, тыс. кв.м.')
plt.ylim(0, 10000)

plt.bar_label(ax, padding=5)

plt.show()
```

```python
plt.figure(figsize=(8, 6))

plt.barh(df['Год'], df['Торговые площади Магнит, тыс. кв.м.'], 
         height=0.5, 
         color='red', 
         alpha=0.5)

plt.title('Торговые площади ПАО "Магнит"')
plt.xlabel('Год')
plt.ylabel('Торговые площади, тыс. кв.м.')

plt.show()
```

```python
#пример графика для двух групп данных

plt.figure(figsize=(8, 6))

plt.bar(df['Год'], df['Торговые площади Магнит, тыс. кв.м.'],
        color='red', 
        alpha=0.5)

plt.bar(df['Год'], df['Торговые площади X5Group, тыс. кв.м.'], 
        color='green', 
        alpha=0.5)

plt.legend(['Магнит', 'X5Group'])
plt.title('Торговые площади')
plt.xlabel('Год')
plt.ylabel('Торговые площади, тыс. кв.м.')

plt.show()
```

```python
#Решение проблемы наложения двух графиков друг на друга

plt.figure(figsize=(8, 6))

w = 0.1
x = df.index.values

plt.bar(x-w, df['Торговые площади Магнит, тыс. кв.м.'],
        width=w*2, 
        color='red', 
        alpha=0.5)

plt.bar(x+w, df['Торговые площади X5Group, тыс. кв.м.'], 
        width=w*2, 
        color='green', 
        alpha=0.5)

plt.xticks(ticks=x, labels=df['Год'])

plt.legend(['Магнит', 'X5Group'])
plt.title('Торговые площади')
plt.xlabel('Год')
plt.ylabel('Торговые площади, тыс. кв.м.')

plt.show()
```

```python
#Построение аналогичного графика с помощью pandas

df.plot(kind='bar', x='Год', 
        xlabel='Год', 
        ylabel='Торговые площади, тыс. кв.м.',
        figsize=(8, 6), 
        color=['red', 'green'], 
        alpha=0.5, 
        rot=0,
        title='Торговые площади')
```
