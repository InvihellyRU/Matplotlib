Пример в объектно - ориентированном подходе построения графиков

```python
x = np.linspace(-np.pi, np.pi, 20)
cos_x = np.cos(x) 
sin_x = np.sin(x) 

fig = plt.figure(figsize=(7,5))
ax = fig.add_axes([0, 0, 1, 1])

ax1 = ax.plot(x, cos_x, 
              marker='o', 
              c='c', 
              linewidth=2) 

ax2 = ax.plot(x, sin_x, 
              marker='*', 
              c='r', 
              linewidth=2)

ax.legend(labels=['cos(x)', 'sin(x)']) 
ax.grid(alpha=0.3)

ax.set_title('Графики тригонометрических функций', 
             fontsize=15, 
             fontweight='bold', 
             color='c', 
             pad=15)

plt.show()
```
