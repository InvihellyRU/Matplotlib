Параметр color принимает названия на английском языке, например: 'white', 'black', 'yellow', 'orange', 'red, 'green', 'blue' и др. Также можно использовать названия тёмных или светлых оттенков цветов, например: 'darkgreen', 'darkred', 'lightyellow', 'lightgrey'. Есть еще возможность использовать различные оттенки стандартных цветов, с такими названиями как: 'snow', 'silver', 'gold', 'olive', 'lime', 'coral', 'royalblue' и др.

Несколько стандартных цветов можно использовать в сокращённом варианте: 'k' - чёрный, 'w' - белый, 'r' - красный, 'y' - жёлтый, 'g' - зелёный, 'b' - синий, 'c' - цвет морской волны, 'm' - пурпурный.

Аргументы marker и label функции plot не могут принимать список значений для разных линий в одном вызове plt.plot. Чтобы задать разные маркеры для каждой линии, нужно вызвать plt.plot для каждой линии отдельно.

Параметры для loc легенды графика

'best' или 0
'upper right'  или 1
'upper left'  или 2
'lower left' или 3
'lower right' или 4
'right' или 5
'center left' или 6
'center right' или 7
'lower center' или 8
'upper center' или 9
'center' или 10

или loc=[0.5, 0.5] - относительные координаты, каждая из которых может быть от 0 до 1. Если относительная координата больше 1 или меньше 0, то легенда выносится за график

```python
values = [['Австралия', 25], ['Азия', 4540], ['Африка', 1280],
          ['Европа', 742], ['Северная Америка', 543], ['Южная Америка', 428]]
columns = ['Часть света', 'Численность в 2018 г., млн чел.']
df = pd.DataFrame(values, columns=columns)

display(df)
```

Реализация обводки букв в заголовке

```python
fig, ax = plt.subplots()
ax.plot([0, 2], [0, 2])
title = ax.set_title('Заголовок графика', 
                     fontsize=20, 
                     color='white')

title.set_path_effects([PathEffects.withStroke(linewidth=3, foreground='black')])
plt.show()
```

```python
plt.plot(df['Часть света'].values, df['Численность в 2018 г., млн чел.'].values)
plt.grid(color='grey', linestyle='--', linewidth=0.5)

plt.xlabel('Часть света', 
           fontsize = 10)

plt.ylabel('Численность населения, млн чел.', 
           fontsize = 10, 
           fontweight = 'bold')

plt.xticks(rotation = 45)

plt.title('График зависимости численности населения от части света', 
          fontsize = 10, 
          color = 'r', 
          alpha = 0.8)

plt.show()
```

Реализация отображения легенды для barplot

```python
cities = ['City_A', 'City_B', 'City_C', 'City_D']
populations = [1000, 1100, 1200, 1300]
fig, ax = plt.subplots()
bars = ax.bar(cities, populations)
for i, bar in enumerate(bars):
    bar.set_label(f'{cities[i]}: {populations[i]}')
ax.legend(title='Cities')
plt.legend(loc=(1.1, 0.75))
plt.show()
```

Subplots

```python
x = np.linspace(0, 10, 5)
y1 = x**2
y2 = x**3
y3 = x**0.5
y4 = np.sin(x**2)
```

```python
plt.subplot(2, 2, 1) # (количество строк, количество столбцов, индекс конечного графика)
plt.plot(x, y1, 'r') 
plt.grid()
plt.title('y1')

plt.subplot(2, 2, 2)
plt.plot(x, y2, 'g')
plt.grid()
plt.title('y2')

plt.subplot(2, 2, 3)
plt.plot(x, y3, 'y')
plt.grid()
plt.title('y3')

plt.subplot(2, 2, 4)
plt.plot(x, y4, 'c')
plt.grid()
plt.title('y4')

plt.suptitle('Графики') # общие заголовок для всех графиков

plt.tight_layout() # решает проблему наложения графиков друг на друга

plt.show()
```
