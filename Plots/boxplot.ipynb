Диаграмма размаха (boxplot)

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import matplotlib.patheffects as PathEffects
from matplotlib.ticker import AutoMinorLocator
from matplotlib.gridspec import GridSpec
from matplotlib import colormaps

#from sklearn.datasets import load_iris
```

```python
df = pd.read_csv('brent_m_2008_2023.csv')
df
```

```python
plt.figure(figsize=(6, 7))

plt.boxplot(df['high-low'])

plt.ylabel('Разность максимальной и минимальной цен за месяц', 
           labelpad=15)

plt.show()
```

```python
plt.figure(figsize=(7, 5))

ax = plt.boxplot(df[['high', 'low']],
                 patch_artist=True,
                 labels=['high', 'low'])

ax['boxes'][0].set_facecolor('pink')
ax['boxes'][1].set_facecolor('lightgreen')

plt.show()
```

```python
plt.figure(figsize=(7, 5))

ax = plt.boxplot(df[['high', 'low']],
                 patch_artist=True,
                 labels=['high', 'low'])

colors = ['pink', 'lightgreen']
for patch, color in zip(ax['boxes'], colors):
    patch.set_facecolor(color)

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(7, 5))

ax.boxplot(df[['high', 'low']],
           labels=['high', 'low'])
    
ax.yaxis.grid(True)

plt.show()
```
