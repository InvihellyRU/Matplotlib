График гистограммы (hist plot)

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import matplotlib.patheffects as PathEffects
from matplotlib.ticker import AutoMinorLocator
from matplotlib.gridspec import GridSpec
from matplotlib import colormaps
```

```python
np.random.seed(12)

arr = np.random.randn(1000)
```

```python
plt.hist(x=arr, 
         bins=30, 
         edgecolor='black', 
         color='#B1B176')

plt.show()
```

```python
plt.hist(x=arr, 
         histtype='step', 
         color='#B1B176', 
         linewidth=5, 
         bins = 30)

plt.show()
```

```python
plt.hist(x=arr, 
         edgecolor='black', 
         color='#B1B176', 
         orientation='horizontal', 
         bins = 30)

plt.show()
```
