Круговая диаграмма (pie plot)

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import matplotlib.patheffects as PathEffects
from matplotlib.ticker import AutoMinorLocator
from matplotlib.gridspec import GridSpec
from matplotlib import colormaps
```

```python
df_values = [['Цех №1', 300],
             ['Цех №2', 210],
             ['Цех №3', 157],
             ['Цех №4', 180],
             ['Цех №5', 98]]
df_columns = ['Цех', 'Произведено деталей, шт']

df = pd.DataFrame(df_values, columns=df_columns)

display(df)
```

```python
plt.pie(df['Произведено деталей, шт'],
        labels=df['Цех'],
        wedgeprops={'edgecolor': 'w', 'linewidth': 3, 'width': 0.6})

plt.show()
```

```python
explode = [0, 0, 0, 0, 0.1]

plt.pie(df['Произведено деталей, шт'],
        labels=df['Цех'],
        wedgeprops={'edgecolor': 'w'},
        shadow=True, #тень у диаграммы
        explode=explode,
        autopct='%.2f') #выдвигаем одну из частей, передавая список, на сколько % сместить долю

plt.show()
```

```python
explode = [0, 0, 0, 0, 0.2]

plt.pie(df['Произведено деталей, шт'],
        labels=df['Цех'],
        wedgeprops={'edgecolor': 'w', 'linewidth': 3},
        shadow=True, #тень у диаграммы
        explode=explode, #выдвигаем одну из частей, передавая список, на сколько % сместить долю
        startangle=3,
        autopct='%1.1f%%') # делаем поворот против часовой стрелки (+) или по часовой стрелке (-)
plt.show()
```

```python
colors = ['#B1B176', '#DBA4F6', '#B7DC82', '#A4F6CB', '#C5CCAD']

plt.pie(df['Произведено деталей, шт'],
        labels=df['Цех'],
        wedgeprops={'edgecolor': 'w', 'width': 0.4, 'linewidth': 4},
        colors=colors,
        radius=1.15,
        autopct='%1.1f%%',
        pctdistance=0.82,
        labeldistance=1.15)
plt.show()
```
